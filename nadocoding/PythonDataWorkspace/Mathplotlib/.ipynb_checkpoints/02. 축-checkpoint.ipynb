```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'AppleGothic'
#matplot.rcParams['font.family'] = 'Malgun Gothic' 윈도우

matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = []
for i in range(-100,100):
    x.append(i)
y = np.square(x)
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold':'size':20})
# 개별 폰트 설정
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
plt.plot(x,y)
plt.xlabel('X axis', color='red')
plt.ylabel('Y axis', color='#00aa00')
```

```python
plt.plot(x,y)
plt.xlabel('X axis', color='red',loc='right') # left center right
plt.ylabel('Y axis', color='#00aa00',loc='top') # top center bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([1,4,9,16])
plt.show()
```
