# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
import numpy as np

x = []
for i in range(-100,100):
    x.append(i)
y = np.square(x)

plt.plot(x,y)
```

```python
print(plt.plot(x,y)) #리턴되는 결과가 글자로 나옴
```

```python
plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title('Line Graph')
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프')
```
