# 3. 범례(legend)

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'AppleGothic'
#matplot.rcParams['font.family'] = 'Malgun Gothic' 윈도우

matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y, label='무슨데이터')
plt.legend()
```

```python
plt.plot(x,y, label='무슨데이터')
plt.legend(loc='upper right')
```

```python
plt.plot(x,y, label='무슨데이터')
plt.legend(loc='lower right')
```

```python
plt.plot(x,y, label='무슨데이터')
plt.legend(loc='best')
```

```python
plt.plot(x,y, label='범례')
plt.legend(loc=(0.5,0.5)) #x axis, y axis (0~1 사이)
```

```python
plt.plot(x,y, label='범례')
plt.legend(loc=(0.2,0)) #x axis, y axis (0~1 사이)
```
