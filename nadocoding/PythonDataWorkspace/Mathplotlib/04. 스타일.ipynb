# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'AppleGothic'
#matplot.rcParams['font.family'] = 'Malgun Gothic' 윈도우

matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x=[1,2,3]
y=[2,4,8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth=5)
```

## Marker

```python
plt.plot(x,y,marker='o')
```

```python
plt.plot(x,y,marker='o',linestyle='None')
```

```python
plt.plot(x,y,marker='v')
```

```python
plt.plot(x,y,marker='v',markersize=10)
```

```python
plt.plot(x,y,marker='x',markersize=10)
```

### matplotlib.markers 라고 구글 검색하면 다 나옴

```python
plt.plot(x,y,marker='o',markersize=10, markeredgecolor='red')
```

```python
plt.plot(x,y,marker='o',markersize=10, markeredgecolor='red',markerfacecolor='yellow')
```

## 축약어
markerfacecolor = mfc
markersize = ms
markeredgecolor = mec
linestyle = ls

```python
plt.plot(x,y,marker='o',mfc='red',ms=10,mec='blue',ls=':')
```

#### ㄴ matplotlib.pyplot.plot

## 투명도

```python
plt.plot(x,y,marker='o',mfc='red',ms=10,mec='blue',ls=':',alpha=0.3)
# alpha: 투명도 (0~1)
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5),dpi=50) #dpi 해상도 dots per inch
plt.plot(x,y)
```

## 배경 색

```python
plt.figure(facecolor='#0000aa')
plt.plot(x,y)
```
