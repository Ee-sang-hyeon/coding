# 9. 막대 그래프 (심화)

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'AppleGothic'
#matplot.rcParams['font.family'] = 'Malgun Gothic' 윈도우

matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호','서태웅','정대만'] 
values = [190,187,184]

plt.barh(labels,values)
```

```python
plt.barh(labels,values)
plt.ylim(175,190)
```

```python
plt.barh(labels,values)
plt.xlim(175,195)
```

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('/') #/// shape
bar[1].set_hatch('x') #xxx
bar[2].set_hatch('...') #...
```

#### ㄴ matplotlib hatch 구글링

```python
bar = plt.bar(labels,values)
bar[0].set_hatch('||') #/// shape
bar[1].set_hatch('*') #xxx
bar[2].set_hatch('//') #...

```

```python
bar = plt.bar(labels, values)
plt.ylim(175,195)

for idx, rect in enumerate(bar):
    plt.text(idx,rect.get_height()+0.5,values[idx], ha='center', color='blue')
```
