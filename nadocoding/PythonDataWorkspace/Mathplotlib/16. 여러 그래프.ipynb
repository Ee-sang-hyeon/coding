# 16. 여러 그래프

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'AppleGothic'
#matplot.rcParams['font.family'] = 'Malgun Gothic' 윈도우

matplotlib.rcParams['font.size'] = 15 #글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
# 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
import pandas as pd
df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,10)) #2x2 크기의 그래프 행렬
fig.suptitle('여러 그래프 넣기')
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,10)) #2x2 크기의 그래프 행렬
fig.suptitle('여러 그래프 넣기')

axs[0,0].bar(df['이름'],df['국어'],label='국어점수') #데이터 설정
axs[0,0].set_title('첫 번째 그리프') # 제목
axs[0,0].legend() # 범례
axs[0,0].set(xlabel='이름',ylabel='점수') # x, y 축 label
axs[0,0].set_facecolor('lightyellow') # 배경 색
axs[0,0].grid(linestyle='--', linewidth=0.5)
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,10)) #2x2 크기의 그래프 행렬
fig.suptitle('여러 그래프 넣기')

#첫번째 그래프
axs[0,0].bar(df['이름'],df['국어'],label='국어점수') #데이터 설정
axs[0,0].set_title('첫 번째 그리프') # 제목
axs[0,0].legend() # 범례
axs[0,0].set(xlabel='이름',ylabel='점수') # x, y 축 label
axs[0,0].set_facecolor('lightyellow') # 배경 색
axs[0,0].grid(linestyle='--', linewidth=0.5)

#두번째 그래프
axs[0,1].plot(df['이름'],df['수학'],label='수학')
axs[0,1].plot(df['이름'],df['영어'],label='수학')
axs[0,1].legend()

#세번째 그래프
axs[1,0].barh(df['이름'],df['키'])

#네번째 그래프
axs[1,1].plot(df['이름'],df['사회'],color='g',alpha=0.5)

plt.savefig('severalgraph.png')
```
