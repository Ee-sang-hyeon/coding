# 12. 데이터 수정

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## Column 수정

```python
df['학교'].replace({'북산고':'상북고','능남고':'무슨고'})
```

```python
df
```

```python
df['학교'].replace({'북산고':'상북고','능남고':'무슨고'},inplace=True)
df
```

```python
df['SW특기'].str.lower()
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df['SW특기'] = df['SW특기'].str.upper()
df
```

```python
df['학교'] = df['학교'] + '등학교' # 학교 데이터 + 등학교
df
```

## Column 추가

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df
```

```python
df['result'] = 'Fail' #결과 column 을 추가하고 전체 데이터는 fail로 초기화
df
```

```python
df.loc[df['총합']>400,'result'] = 'Pass' # 총합이 400보다 큰 데이터에 대해서 결과를 pass로 업데이트
df
```

## Column 삭제

```python
df.drop(columns=['총합']) #총합 column 을 삭제
```

```python
df.drop(columns=['총합','국어','영어']) # 국어, 영어, 수학 column을 삭제
```

## Row 삭제

```python
df.drop(index='4번') #4번 학생 데이터 row 삭제
```

```python
filt = df['수학'] < 80 # 수학 점수 80점 미만 학생 필터링
df[filt]
```

```python
df[filt].index
```

```python
df.drop(index=df[filt].index) #필터 조건에 맞는 행을 삭제
```

```python
df # inplace 해줘야함
```

## Row 추가

```python
df.loc['9번'] = ['이정환','해남고등학교',184,90,90,90,90,90,'Kotlin',450,'Pass']
```

```python
df
```

## Cell 수정

```python
df.loc['4번','SW특기'] = 'Python' # 4번 학생의 솦 특기를 파이썬으로 변경
df
```

```python
df.loc['5번',['학교','SW특기']] = ['능남고등학교','C'] # 5번 학교는 능남고, 솦 특기는 C로 변경
```

```python
df
```
