# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 조건

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키']>= 185 # 학생들의 키가 185 이상인지 여부를 True/False
```

```python
filt = (df['키']>=185) #필터를 만들어서
df[filt] # 적용
```

```python
df[~filt] #filter를 역으로 적용
```

```python
df[df['키']>=185]
```

```python
df.loc[df['키']>=185,'수학'] # 키가 185이상인 학생들의 수학점수
```

```python
df.loc[df['키']>=185,['수학','과학']] #키가 185이상인 학생들의 수학, 과학점수
```

## 다양한 조건
### &그리고

```python
df.loc[(df['키']>=185) & (df['학교']=='북산고')] #키가 185 이상이고 북산고 다니는 학생의 데이터
```

### | 또는 

```python
df.loc[(df['키']<170) | (df['키']>200)]
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # '송'씨 성을 가진 사람
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태' 가. 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름에 '태'가 안들어가는 사람
```

```python
langs = ["Python", "Java"]
filt = df['SW특기'].isin(langs) #특기가 python과 java인 학생 필터
df[filt]
```

```python
df #황태산은 대문자로 적혀서 안뽑힘
```

```python
langs = ["python","java"]
filt = df['SW특기'].str.lower().isin(langs) # 다 소문자로 바꿔서 찾음
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
df[filt] #NaN data 처리 안됨.
```

```python
df['SW특기'].str.lower().isin(langs)
```

```python
df['SW특기'].str.contains('Java', na=True) #NaN 데이터에 대해서 True로 설정
```

```python
df['SW특기'].str.contains('Java', na=False) 
# NaN data 를 False로 설정
```

```python
filt = df['SW특기'].str.contains('Java', na=False)
df[filt] # data를 잘 가져온다.
```
