# 다항회기분석. Polynomial Regression
Simple Regression은 두 변수 간의 관계를 직선 형태로 설명하는 알고리즘이었다. 따라서 직선보다는 곡선으로 설명하는 것이 적합할 때는 단순회귀분석 보다는 다항함수를 사용하여 복잡한 곡선 형태의 회귀선을 표현할 수 있다.
즉, 다항회귀분석은 2차함수 이상의 다항 함수를 이용하여 두 변수 간의 선형관계를 설명하는 알고리즘이다.

$$Y = aX^2 + bX + c$$

학습을 통해 3개의 계수 a,b,c를 찾아서 모형을 완성한다

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('auto-mpg.csv',header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','name']

df['horsepower'].replace('?',np.nan, inplace=True) # ?를 nan으로 변경
df.dropna(subset=['horsepower'], axis=0, inplace=True) # 누락데이터 행을 삭제
df['horsepower']=df['horsepower'].astype('float') # 문자열을 실수형으로 변환

ndf = df[['mpg','cylinders','horsepower','weight']]

X=ndf[['weight']]
y=ndf['mpg']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=10)
```

다항회귀분석을 사용하기 위해서는 선형회귀분석에서 사용한 LinearRegression을 같게 사용하지만, PolymonialFeatures를 활용하여, 학습에서 사용하는 X값들을 2차원으로 바꾼 후에 fit에 넣어준다.

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures # 다항함수 변환
# preprocessing -> 전처리
# 전 처리 중에 PolynomialFeatures는 비선형 feature에 복잡한 계산을 알아서 해준다.

poly = PolynomialFeatures(degree=2) # degree = 차원
X_train_poly = poly.fit_transform(X_train) # X-train을 다항식으로 변환할 수 있도록 fit_transform을 해줌

print('원래 데이터: ', X_train.shape)
print('2차항 변환 데이터: ', X_train_poly.shape)
```

ㄴ X_tarin의 1개 열이 X_train_poly에서는 3개 열로 늘어난다.

#### fit 메소드를 이용해 학습 시작

```python
pr = LinearRegression()
pr.fit(X_train_poly, y_train)
```

#### score 메소드로 모형의 결정계수(R-제곱)을 구한다.
이때 X_test도 fit_transform 메소드를 활용하여 2차항 회귀분석에 맞게 변환해야한다.

```python
X_test_poly = poly.fit_transform(X_test) # X test 변환
r_square = pr.score(X_test_poly, y_test)
print('이것의 성능은 ',r_square)
```

위에서 70정도의 정확도가 나온 것을 알 수 있다.

#### 회귀선 그래프로 출력
- y_hat_test의 산점도와 test data로 예측한 회귀선을 그래프로 출력하자.
- predict() 메소드에 입력하여 예측한 결과인 y_hat_test를 빨간 점으로 표시하면 회귀선이 된다.

```python
y_hat_test = pr.predict(X_test_poly)

fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1,1,1) # ?
ax.plot(X_train, y_train, 'o', label='Train Data')
ax.plot(X_test, y_hat_test, 'r+', label='Predicted Value')
ax.legend(loc='best')

plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

#### 실제값과 예측값 비교
Seaborn 라이브러리의 distplot() 함수를 사용하여 실제 값 y와 예측값 y_hat의 분포 차이를 비교한다.

```python
X_poly = poly.fit_transform(X)
y_hat = pr.predict(X_poly)

plt.figure(figsize=(10,5))
ax1 = sns.distplot(y, hist=False, label="y")
ax2 = sns.distplot(y_hat, hist=False, label='y_hat',ax=ax1)

plt.show()
```

단순 회기분석 보델보다 편향되는 경향이 감소된 것으로 볼 때, 더 적합한 모형이라고 할 수 있다.
